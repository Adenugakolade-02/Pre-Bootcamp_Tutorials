# Conditional Statements in Python

There are majorly three conditional statements in python and they are;

**1. if Statement**

**2. if else Statement**

**3. elif Statement**

**4. else Statement**

Majority of the time when programming, we'll need to control the flow of our logic. Our program will want to perform an action in only certain cases, we can use the **if**, **elif**, and **else** statements to control for these cases.
### if Statement
**if** statement is a programming coditional statement that if proven true it performs a function or displays an information.

**Let us check out some of the examples below**



## Simple if Statement

    if some_condition:
        # Do Something


```python
if taiwo<shola:
    print('Taiwo is lesser than Shola')
    
```

```python
if shola>taiwo:
    print("Shola is greater than Taiwo")
```

## if else Statement

Let's now add in an alternate action in case the **if** is not True using the **else** statement.

Notice the format and how the code blocks line up, this is crucial in Python! Code indentation let's Python know what blocks and statements correspond together.

Note: Always take note of identation as it could cause a major disaster while running your code though it is easy to spot

```python
if 1==1:
    print("One is equal to One")
else:
    print("First if was not True")
```

```python
if 1==2:
    print("One is equal to Two")
else:
    print("First if was not True")
```

```python
saved_password = 123456
new_password = input("Enter your password")

if int(new_password) == saved_password:
    print("Password correct")
else:
    print("You're a criminal")
```

## elif Statement

Now let's imagine we have multiple conditions to check before the final **else** statement, this is where we can use the **elif** keyword to check for as many individual conditions as necessary:

```python
if 2 == 0:
    print('First condition True')
elif 2 == 1:
    print("Second condition True")
elif 2 == 100:
    print("Third condition True")
else:
    print("None of the above conditions were True")
```

Let's see what happens if we had multiple True conditions:

```python
if 2 == 2:
    print('First condition True')
elif 2 == 2:
    print("Second condition True")
elif 2 == 2:
    print("Third condition True")
else:
    print("None of the above conditions were True")
```

Notice how only the first True condition's code block will be executed, keep this in mind when writing your scripts!
